```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
ride_data = pd.merge(ride_data,city_data,on='city',how='left')

# Display the data table for preview
ride_data
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

#Rides By City
number_of_rides_by_city = ride_data.groupby(['city','type'])
number_of_rides_by_city = pd.DataFrame(number_of_rides_by_city['ride_id'].count())
number_of_rides_by_city = number_of_rides_by_city.rename(columns = {'ride_id': 'ride_id_count'}) 
number_of_rides_by_city = number_of_rides_by_city.reset_index()

#Fares By City
average_fare_by_city = ride_data.groupby(['city','type'])
average_fare_by_city = pd.DataFrame(average_fare_by_city['fare'].mean())
average_fare_by_city = average_fare_by_city.rename(columns = {'fare':'fare_mean'})
average_fare_by_city = average_fare_by_city.reset_index()

#Driver Count and Type  By City
driver_count_and_type_by_city = city_data.sort_values(by=['city'])
driver_count_and_type_by_city = driver_count_and_type_by_city[['city','driver_count','type']]

#Build the scatter plots for each city types

#Urban
urban_rides = number_of_rides_by_city.loc[number_of_rides_by_city['type'] == 'Urban','ride_id_count'].values
urban_fares = average_fare_by_city.loc[average_fare_by_city['type'] == 'Urban','fare_mean'].values
urban_driver_count = driver_count_and_type_by_city.loc[driver_count_and_type_by_city['type'] == 'Urban','driver_count'].values
urban_plot = plt.scatter(urban_rides,urban_fares,marker = 'o',c = 'coral',edgecolor= 'black',s = urban_driver_count*10, alpha = 0.75,label = 'Urban')

#Suburban
suburban_rides = number_of_rides_by_city.loc[number_of_rides_by_city['type'] == 'Suburban','ride_id_count'].values
suburban_fares = average_fare_by_city.loc[average_fare_by_city['type'] == 'Suburban','fare_mean'].values
suburban_driver_count = driver_count_and_type_by_city.loc[driver_count_and_type_by_city['type'] == 'Suburban','driver_count'].values
suburban_plot = plt.scatter(suburban_rides,suburban_fares,marker = 'o',c = 'skyblue',edgecolor= 'black',s = suburban_driver_count*10, alpha = 0.75,label = 'Suburban')

#Rural
rural_rides = number_of_rides_by_city.loc[number_of_rides_by_city['type'] == 'Rural','ride_id_count'].values
rural_fares = average_fare_by_city.loc[average_fare_by_city['type'] == 'Rural','fare_mean'].values
rural_driver_count = driver_count_and_type_by_city.loc[driver_count_and_type_by_city['type'] == 'Rural','driver_count'].values
rural_plot = plt.scatter(rural_rides,rural_fares,marker = 'o',c = 'gold',edgecolor= 'black',s = rural_driver_count*10, alpha = 0.75,label = 'Rural')

plt.plot()

#Incorporate the other graph properties
plt.grid()
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.title('Pyber Ride Sharing Data (2016)')

#Create a legend
lgnd = plt.legend(loc='best')
lgnd.legendHandles[0]._sizes = [100]
lgnd.legendHandles[1]._sizes = [100]
lgnd.legendHandles[2]._sizes = [100]

#Incorporate a text label regarding circle size
plt.text(42,40,'Note:')
plt.text(42,38,'Circle size correlates with driver count per city.')
#Save Figure
plt.savefig('pyber_plot_city_type_vs_avg_fare.jpg')

# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
fares_by_city_type = ride_data.groupby('type')
fares_by_city_type = pd.DataFrame(fares_by_city_type['fare'].sum())
fares_by_city_type['fare percent'] = fares_by_city_type['fare']/fares_by_city_type['fare'].sum()

# Build Pie Chart
labels = ['Rural','Suburban','Urban']
values = fares_by_city_type['fare percent'].values
colors = ['gold','skyblue','coral']
explode = [0,0,0.1]
plt.pie(values,explode=explode,labels=labels, colors=colors,autopct="%1.1f%%",shadow=True,startangle=180,pctdistance=0.6)
plt.axis('auto')
plt.title('% of Total Fares By City Type')

# Save Figure
plt.savefig('pyber_plot_city_type_vs_fare_percent.jpg')

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
rides_by_city_type = ride_data.groupby('type')
rides_by_city_type = pd.DataFrame(rides_by_city_type['ride_id'].count())
rides_by_city_type['ride percent'] = rides_by_city_type['ride_id']/rides_by_city_type['ride_id'].sum()

# Build Pie Chart
labels = ['Rural','Suburban','Urban']
values = rides_by_city_type['ride percent'].values
colors = ['gold','skyblue','coral']
explode = [0,0,0.1]
plt.pie(values,explode=explode,labels=labels, colors=colors,autopct="%1.1f%%",shadow=True,startangle=180,pctdistance=0.6)
plt.axis('auto')
plt.title('% of Total Rides By City Type')

# Save Figure
plt.savefig('pyber_plot_city_type_vs_ride_percent.jpg')

# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
drivers_by_city_type = city_data.groupby('type')
drivers_by_city_type = pd.DataFrame(drivers_by_city_type['driver_count'].sum())
drivers_by_city_type['driver percent'] = drivers_by_city_type['driver_count']/drivers_by_city_type['driver_count'].sum()

# Build Pie Charts
labels = ['Rural','Suburban','Urban']
values = drivers_by_city_type['driver percent'].values
colors = ['gold','skyblue','coral']
explode = [0,0,0.1]
plt.pie(values,explode=explode,labels=labels, colors=colors,autopct="%1.1f%%",shadow=True,startangle=180,pctdistance=0.6)
plt.axis('auto')
plt.title('% of Total Drivers By City Type')

# Save Figure
plt.savefig('pyber_plot_city_type_vs_driver_percent.jpg')

# Show Figure
plt.show()
```

Trend Number 1:
The average fare is higher in rural areas, lower in suburbs, and lower still in urban areas.  This could be due to differences in ride distance, as rural areas are more spread out than urban ones.

Trend Number 2:
Driver counts in urban areas are higher than suburbs and rural areas, as one might expect with higher population in cities.  But driver counts are not linear with rides or fares.  Cities possess 81% of drivers but just 63% of fares, suggesting higher competition for fares even with more potential customers.

Trend Number 3:
This disparity becomes even more apparent when comparing fares per driver in the 3 city types.  Rural drivers commanded 55 dollars of fares per driver, while Suburban drivers had 40 dollars and urban drivers just 17 dollars. (See Below)
